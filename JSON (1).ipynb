```python
import json

print("JSON Refiner Advanced Edition")

# -------- JSON Validation --------
def validate_json(data):
    try:
        json.dumps(data)
        return True
    except Exception as e:
        print("Invalid JSON:", e)
        return False


# -------- Flatten JSON --------
def flatten_json(data, parent_key='', sep='_'):

    items = {}

    for key, value in data.items():

        new_key = parent_key + sep + key.lower() if parent_key else key.lower()

        if isinstance(value, dict):
            items.update(flatten_json(value, new_key, sep=sep))

        else:

            # Automatic type inference
            if isinstance(value, str) and value.isdigit():
                value = int(value)

            items[new_key] = value

    return items


# -------- JSON Refiner --------
def json_refiner(input_json):

    if not validate_json(input_json):
        return None

    refined = flatten_json(input_json)

    return refined


# -------- USER INPUT --------
print("\nPaste your JSON input below:\n")

user_input = input()

try:
    json_data = json.loads(user_input)

    print("\nInput JSON:\n")
    print(json.dumps(json_data, indent=4))

    result = json_refiner(json_data)

    print("\nRefined JSON Output:\n")
    print(json.dumps(result, indent=4))

except:
    print("Invalid JSON format. Please enter correct JSON.")
```
